```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTENC
```

```python
credito_df_obj = pd.read_csv("credit_simple.csv", sep = ";")
credito_df_obj.head(6)
```

```python
count = credito_df_obj.groupby(["CLASSE"]).size()
count
```

```python
y = credito_df_obj["CLASSE"].values
x = credito_df_obj.iloc[:,:-1].values
```

```python
labelencoder_inst = LabelEncoder()
for i in range(x.shape[1]):
  if x[:,i].dtype == 'object':
    x[:,i] = labelencoder_inst.fit_transform(x[:,i])
```

```python
sm_mod_obj = SMOTENC(random_state = 0, categorical_features = [3,5,6])
x_res, y_res = sm_mod_obj.fit_resample(x,y)
```

```python
x_res
```

```python
df_obj2 = pd.DataFrame({"CLASSE":y_res})
df_obj2.value_counts()
```
