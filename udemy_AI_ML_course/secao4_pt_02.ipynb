```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
iris = datasets.load_iris()
previsores = iris.data
classe = iris.target
previsores
```

```python
sc_mod_obj = StandardScaler()
previsores = sc_mod_obj.fit_transform(previsores)
previsores
```

```python
previsores
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, classe, test_size = 0.3, random_state = 1234)
```

```python
floresta_mod_obj = RandomForestClassifier(n_estimators = 100, random_state = 123)
floresta_mod_obj.fit(x_treino, y_treino)
previsores = floresta_mod_obj.predict(x_teste)
taxa_acerto = accuracy_score(y_teste, previsores)
taxa_acerto
```

```python
pca_mod_obj = PCA(n_components = 3)
previsores = pca_mod_obj.fit_transform(x_teste)
previsores
```
