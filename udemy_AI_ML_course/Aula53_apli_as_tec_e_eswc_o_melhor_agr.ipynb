```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
```

```python
def compare_algorithms(x, max_clusters):
  results = []
  cluster_range = range(2, max_clusters+1)

  #kmeans
  for n_clusters in cluster_range:
    kmeans_obj = KMeans(n_clusters = n_clusters, random_state = 0, n_init = 'auto')
    clusters = kmeans_obj.fit_predict(x)
    silhouette_avg = silhouette_score(x, clusters)
    results.append(("KMeans", n_clusters, silhouette_avg))

  #agglomerative
  for n_clusters in cluster_range:
    agglomerative_obj = AgglomerativeClustering(n_clusters = n_clusters)
    clusters = agglomerative_obj.fit_predict(x)
    silhouette_avg = silhouette_score(x, clusters)
    results.append(('KMeans', n_clusters, silhouette_avg))

  #dbscan
  eps_values = np.arange(0.1, 0.9, 0.1)
  for eps in eps_values:
    dbscan_obj = DBSCAN(eps,min_samples = 5)
    clusters = dbscan_obj.fit_predict(x)
    if len(set(clusters))> 1:
      silhouette_avg = silhouette_score(x, clusters)
      results.append(("DBSCAN", eps, silhouette_avg))
  return results
```

```python
##Escolhendo o melhor agrupador [aula 54, cont]
```

```python
iris = datasets.load_iris()
scaler_obj = StandardScaler()

scaled_data = scaler_obj.fit_transform(iris.data)

results = compare_algorithms(scaled_data, 10)
df_obj = pd.DataFrame(results, columns = ["Agrupador","Clusters","Score"])
#print(results)
df_obj.head() #o impoertante eh o score [a metrica mais imp]
```
