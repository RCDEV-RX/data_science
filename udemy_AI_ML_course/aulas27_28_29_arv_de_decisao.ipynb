#Arvores de Decisao [modelo white-box]

```python
import pandas as pd
from sklearn.model_selection import train_test_split
#from sklearn.naive_bayes import GaussianNB #o modelo de Naive_Bayes
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score,  f1_score, classification_report
#from yellowbrick.classifier import ConfusionMatrix #nao usaremos a confusion matrix aki
from sklearn.metrics import classification_report
from sklearn.tree import export_graphviz
import graphviz
```

```python
base_inst = pd.read_csv("insurance.csv")
base_inst.head(5)
```

```python
base_inst = base_inst.drop(columns = ["Unnamed: 0"])
base_inst = base_inst.dropna()
base_inst.head()
```

```python
y = base_inst.iloc[:,7].values
x = base_inst.iloc[:, [n for n in range(0,27) if n != 7]].values
```

```python
labelencoder = LabelEncoder()
```

```python
for i in range(x.shape[1]):
  if x[:,1].dtype == "object":
    x[:,i] = labelencoder.fit_transform(x[:,i])
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(
    x, y, test_size = 0.3, random_state = 1
)
```

```python
#model = DecisionTreeClassifier(random_state = 1) ##tem muitohyper-parametro aqui;
#model = DecisionTreeClassifier(random_state = 1, max_depth = 8)
model = DecisionTreeClassifier(random_state = 1, max_depth = 8, max_leaf_nodes = 8)
model.fit(x_treino, y_treino)
```

```python
previsoes = model.predict(x_teste)
```

```python
accuracy = accuracy_score(y_teste, previsoes)
precision = precision_score(y_teste, previsoes, average = 'weighted')
recall = recall_score(y_teste, previsoes, average = 'weighted')
f1 = f1_score(y_teste, previsoes, average = 'weighted')
print(f'accuracy: {accuracy}, precision: {precision}, recall: {recall}, f1: {f1}')
```

```python
report = classification_report(y_teste, previsoes)
print(report)
```

```python
dot_data = export_graphviz(model, out_file = None, filled = True, feature_names = base_inst.columns[:-1], class_names = True, rounded = True)

graph = graphviz.Source(dot_data)
graph.render("decision_tree", format = "png")
graph.view()
```
